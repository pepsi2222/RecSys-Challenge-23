```python
import pandas as pd
import numpy as np
```

```python
tst_data_path = '/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/test.csv'
tst_df = pd.read_csv(tst_data_path, sep='\t')
```

```python
pred_clicked = np.random.choice([0, 1], size=len(tst_df), p=[0.5, 0.5])
pred_clicked = pred_clicked.astype(np.int8)
tst_df['is_clicked'] = pred_clicked
```

```python
pred_installed = np.random.choice([0, 1], size=len(tst_df), p=[0.5, 0.5])
pred_installed = pred_installed.astype(np.int8)
tst_df['is_installed'] = pred_installed
```

```python
tst_df.sample(5)
```

```python
columns = ['f_0', 'is_clicked', 'is_installed']
```

```python
tst_df[columns].to_csv('/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/randpred.csv', index=False, sep='\t')
```

```python
pred_clicked = np.ones(len(tst_df), dtype=np.int8)
pred_installed = np.ones(len(tst_df), dtype=np.int8)
tst_df['is_clicked'] = pred_clicked
tst_df['is_installed'] = pred_installed
tst_df[columns].to_csv('/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/pospred.csv', index=False, sep='\t')
```

```python
pred_clicked =  np.zeros(len(tst_df), dtype=np.int8)
pred_installed = np.zeros(len(tst_df), dtype=np.int8)
tst_df['is_clicked'] = pred_clicked
tst_df['is_installed'] = pred_installed
tst_df[columns].to_csv('/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/negpred.csv', index=False, sep='\t')
```
