```python
import pandas as pd

install_df = {}
click_df = {}

def add_prob(pth, model, post=''):
    df_tmp = pd.read_csv(pth, sep='\t')
    if 'install' in pth:
        if model == 'LightGBM':
            install_df[f'p_install_{model}{post}'] = df_tmp['is_installed_prob'].map(lambda x: 0 if x < 0 else 1 if x > 1 else x) 
        else:
            install_df[f'p_install_{model}{post}'] = df_tmp['is_installed_prob']
    if 'click' in pth:
        if model == 'LightGBM':
            click_df[f'p_click_{model}{post}'] = df_tmp['is_clicked_prob'].map(lambda x: 0 if x < 0 else 1 if x > 1 else x) 
        else:
            click_df[f'p_click_{model}{post}'] = df_tmp['is_clicked_prob']
```

```python
files0 = [
    ("/root/autodl-tmp/xingmei/RecSysChallenge23/predictions/LightGBM/2023-06-20-16-57-39is_installed.csv", 'LightGBM', '_675'),
    ("/root/autodl-tmp/xingmei/RecSysChallenge23/predictions/LightGBM/2023-06-20-15-52-22is_installed.csv", 'LightGBM', '_674'),
    ('/root/autodl-tmp/xingmei/RecSysChallenge23/predictions/CatBoost/2023-06-20-17-33-30is_installed.csv', 'CatBoost', '_662'),
    ('/root/autodl-tmp/xingmei/RecSysChallenge23/predictions/CatBoost/2023-06-20-20-28-27is_installed.csv', 'CatBoost', '_647')
]
```

```python
for pth, model, post in files0:
    add_prob(pth, model, post)
install_df = pd.DataFrame(install_df)
click_df = pd.DataFrame(click_df) 
```
