```python
import pandas as pd
import os

xm_pred_dir_3090 = '/root/autodl-tmp/xingmei/RecSys23/predictions'
xm_pred_dir_1401 = '/root/autodl-tmp/xingmei/RecSys23/predictions1401'
xm_pred_dirs = [xm_pred_dir_3090, xm_pred_dir_1401]


# yk_pred_dir_3090 = ...
```

```python
install_df = {}
click_df = {}
for pred_dir in xm_pred_dirs:
    for model_dir in os.listdir(pred_dir):
        pred_pth = os.path.join(
                        pred_dir, 
                        model_dir, 
                        os.listdir(os.path.join(pred_dir, model_dir))[-1])    # 只取最后一个预测文件
        df_tmp = pd.read_csv(pred_pth, sep='\t')
        if 'install' in pred_pth:
            install_df[f'p_install_{model_dir}'] = df_tmp['is_installed_prob']
        if 'click' in pred_pth:
            click_df[f'p_click_{model_dir}'] = df_tmp['is_installed_prob']
```

```python
install_df = pd.DataFrame(install_df)
click_df = pd.DataFrame(click_df)
```

```python
whole_df = pd.read_csv("/root/autodl-tmp/yankai/data/data/simplified_trn_val_tst.csv", sep='\t')
```

```python
install_df['is_installed'] = whole_df['is_installed']
```

```python
install_df.to_csv('/root/autodl-tmp/xingmei/RecSys23/data/install_probs_trn_val_tst.csv', sep='\t', index=False)
click_df.to_csv('/root/autodl-tmp/xingmei/RecSys23/data/click_probs_trn_val_tst.csv', sep='\t', index=False)
```
