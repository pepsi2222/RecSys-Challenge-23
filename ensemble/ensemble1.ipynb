```python
import pandas as pd

idx = pd.read_csv('/root/autodl-tmp/xingmei/RecSys23/data/index_val536066.csv', sep='\t')

def reset_idx(col):
    df_tmp = pd.DataFrame(col)
    df_tmp['idx'] = idx
    df_tmp.sort_values(by='idx', inplace=True)
    df_tmp.drop(columns=['idx'], inplace=True)
    df_tmp.reset_index(inplace=True)
    name = col.name.replace('_536063', '') if col.name.endswith('_536063') else col.name
    return name, df_tmp[col.name]
```
