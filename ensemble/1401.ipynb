```python
import pandas as pd

install_df = {}
click_df = {}

def add_prob(pth, model, post=''):
    df_tmp = pd.read_csv(pth, sep='\t')
    if 'install' in pth:
        install_df[f'p_install_{model}{post}'] = df_tmp['is_installed_prob']
    if 'click' in pth:
        click_df[f'p_click_{model}{post}'] = df_tmp['is_clicked_prob']
```
