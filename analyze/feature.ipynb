```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trn_data_dir = '/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/raw_data/train'
tst_data_dir = '/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/raw_data/test'
```

```python
trn_df = []
for f in os.listdir(trn_data_dir):
    trn_df.append(
        pd.read_csv(
            os.path.join(trn_data_dir, f),
            sep='\t',
            header=0
            )
        )
trn_df = pd.concat(trn_df)
trn_df.to_csv('/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/data/train.csv', sep='\t', index=False)
```

```python
trn_df.sample(5)
```

```python
tst_df = pd.read_csv(os.path.join(tst_data_dir, '000000000000.csv'), sep='\t', header=0)
```

```python
tst_df.sample(5)
```

## Date(f_1)

```python
plt.hist([trn_df['f_1'], tst_df['f_1']], bins=23, label=['train', 'test'], align='mid')
plt.legend(loc='upper right')
plt.xlabel('f1 Date')
plt.ylabel('Num')
plt.savefig('./f_1.png')
```

## Categorical features(f_2 to f_32)

```python
for i in range(2, 33):
    fig, ax = plt.subplots(1, 2, figsize=(20, 5))
    sns.countplot(data=trn_df, y=f'f_{i}', ax=ax[0])
    sns.countplot(data=tst_df, y=f'f_{i}', ax=ax[1])
    plt.savefig(f'./f_{i}.png')
    plt.show()
    plt.close()

    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null ratio: {trn_df[f'f_{i}'].isnull().mean()} | "
          f"Unique num: {len(trn_df[f'f_{i}'].unique())}"
        )
    print(f"Test  statistic - "
          f"Null ratio: {tst_df[f'f_{i}'].isnull().mean()} | "
          f"Unique num: {len(tst_df[f'f_{i}'].unique())}"
    )
```

## Binary features(f_33 to f_41)

```python
for i in range(33, 42):
    plt.hist([trn_df[f'f_{i}'].astype(str), tst_df[f'f_{i}'].astype(str)], bins=2, label=['train', 'test'], align='mid')
    plt.legend(loc='best')
    plt.xlabel(f'f_{i}')
    plt.ylabel('Num')
    plt.savefig(f'./f_{i}.png')
    plt.show()
    plt.close()
    
    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null: {trn_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Ratio of 1: {trn_df[f'f_{i}'].mean():.2f}"
        )
    print(f"Test  statistic - "
          f"Null: {tst_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Ratio of 1: {tst_df[f'f_{i}'].mean():.2f}"
    )
```

## Numerical features(f_42 to f_79)

```python
for i in range(42, 80):
    plt.hist([trn_df[f'f_{i}'], tst_df[f'f_{i}']], label=['train', 'test'], align='mid')
    plt.legend(loc='best')
    plt.xlabel(f'f_{i}')
    plt.ylabel('Num')
    plt.savefig(f'./f_{i}.png')
    plt.show()
    plt.close()
    
    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null: {trn_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Mean: {trn_df[f'f_{i}'].mean():.2f} | Median {trn_df[f'f_{i}'].median():.2f} | "
          f"Min: {trn_df[f'f_{i}'].min():.2f} | Max {trn_df[f'f_{i}'].max():.2f}"
        )
    print(f"Test  statistic - "
          f"Null: {tst_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Mean: {tst_df[f'f_{i}'].mean():.2f} | Median {tst_df[f'f_{i}'].median():.2f} | "
          f"Min: {tst_df[f'f_{i}'].min():.2f} | Max {tst_df[f'f_{i}'].max():.2f}"
    )
```

## Labels(is_clicked, is_installed)

```python
for i in ['is_clicked', 'is_installed']:
    plt.hist(trn_df[i].astype(str), bins=2, label='train', align='mid')
    plt.legend(loc='best')
    plt.xlabel(f'{i}')
    plt.ylabel('Num')
    plt.savefig(f'./{i}.png')
    plt.show()
    plt.close()
    
    print(f"{i} \n"
          f"Train statistic - "
          f"Null: {trn_df[i].isnull().mean():.2f} | "
          f"Ratio of 1: {trn_df[i].mean():.2f}"
        )
```
