```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

## Raw data

```python
trn_data_dir = '/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/raw_data/train'
tst_data_dir = '/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/raw_data/test'
```

```python
trn_df = []
for f in os.listdir(trn_data_dir):
    trn_df.append(
        pd.read_csv(
            os.path.join(trn_data_dir, f),
            sep='\t',
            header=0
            )
        )
trn_df = pd.concat(trn_df)
trn_df.to_csv('/root/autodl-tmp/xingmei/Sharechat-RecSys-Challenge-23/data/train.csv', sep='\t', index=False)
```

```python
tst_df = pd.read_csv(os.path.join(tst_data_dir, '000000000000.csv'), sep='\t', header=0)
```

## Concated data

```python
trn_data_pth = '/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/train.csv'
tst_data_pth = '/root/data/xingmei/Sharechat-RecSys-Challenge-23/data/test.csv'
trn_df = pd.read_csv(trn_data_pth, sep='\t', header=0)
tst_df = pd.read_csv(tst_data_pth, sep='\t', header=0)
```

## Date(f_1)

```python
plt.hist([trn_df['f_1'], tst_df['f_1']], bins=23, label=['train', 'test'], align='mid')
plt.legend(loc='upper right')
plt.xlabel('f1 Date')
plt.ylabel('Num')
```

## Categorical features(f_2 to f_32)

```python
for i in range(2, 33):
    fig, ax = plt.subplots(1, 2, figsize=(20, 5))
    sns.countplot(data=trn_df, y=f'f_{i}', ax=ax[0])
    sns.countplot(data=tst_df, y=f'f_{i}', ax=ax[1])
    plt.show()
    plt.close()

    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null ratio: {trn_df[f'f_{i}'].isnull().mean()} | "
          f"Unique num: {len(trn_df[f'f_{i}'].unique())}"
        )
    print(f"Test  statistic - "
          f"Null ratio: {tst_df[f'f_{i}'].isnull().mean()} | "
          f"Unique num: {len(tst_df[f'f_{i}'].unique())}"
    )
    tst_ooc = not set(tst_df[f'f_{i}'].unique()).issubset(set(trn_df[f'f_{i}'].unique()))
    if tst_ooc:
        ooc_cat = set(tst_df[f'f_{i}'].unique()) - set(trn_df[f'f_{i}'].unique())
        sum_ = 0
        for k in ooc_cat:
              sum_ += (tst_df[f'f_{i}'] == k).sum()
        print(f'Test is Out-of-category. Num: {len(ooc_cat)} | Ratio: {sum_/len(tst_df[f"f_{i}"])}')        

    unused_cat = set(trn_df[f'f_{i}'].unique()) - set(tst_df[f'f_{i}'].unique())
    if len(unused_cat) > 0:
        sum_ = 0
        for k in unused_cat:
            sum_ += (trn_df[f'f_{i}'] == k).sum()
        print(f'Train exists unused category. Num: {len(unused_cat)} | Ratio: {sum_/len(trn_df[f"f_{i}"])}')
```

## Binary features(f_33 to f_41)

```python
for i in range(33, 42):
    plt.hist([trn_df[f'f_{i}'].astype(str), tst_df[f'f_{i}'].astype(str)], bins=2, label=['train', 'test'], align='mid')
    plt.legend(loc='best')
    plt.xlabel(f'f_{i}')
    plt.ylabel('Num')
    plt.show()
    plt.close()
    
    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null: {trn_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Ratio of 1: {trn_df[f'f_{i}'].mean():.2f}"
        )
    print(f"Test  statistic - "
          f"Null: {tst_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Ratio of 1: {tst_df[f'f_{i}'].mean():.2f}"
    )
```

## Numerical features(f_42 to f_79)

```python
for i in range(42, 80):
    if len(trn_df[f'f_{i}'].unique()) < 50:
        fig, ax = plt.subplots(1, 2, figsize=(20, 5))
        sns.countplot(data=trn_df, y=f'f_{i}', ax=ax[0])
        sns.countplot(data=tst_df, y=f'f_{i}', ax=ax[1])
        plt.show()
        plt.close()
    else:
        fig, ax = plt.subplots(1, 3, figsize=(20, 5))
        ax[0].hist([trn_df[f'f_{i}'], tst_df[f'f_{i}']], bins=10, label=['train', 'test'], align='mid')
        ax[0].legend(loc='best')
        ax[0].set_xlabel(f'f_{i}')
        ax[0].set_ylabel('Num')
        ax[1].hist([trn_df[f'f_{i}'], tst_df[f'f_{i}']], bins=10, label=['train', 'test'], align='mid', 
                   range=[trn_df[f'f_{i}'].min(), 2*trn_df[f'f_{i}'].mean()])
        ax[2].hist([trn_df[f'f_{i}'], tst_df[f'f_{i}']], bins=10, label=['train', 'test'], align='mid', 
                   range=[trn_df[f'f_{i}'].min(), trn_df[f'f_{i}'].mean()])
        plt.show()
        plt.close()
    
    print(f"f_{i} \n"
          f"Train statistic - "
          f"Null: {trn_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Mean: {trn_df[f'f_{i}'].mean():.2f} | Median {trn_df[f'f_{i}'].median():.2f} | "
          f"Min: {trn_df[f'f_{i}'].min():.2f} | Max {trn_df[f'f_{i}'].max():.2f} | "
          f"Unique num: {len(trn_df[f'f_{i}'].unique())}"
        )
    print(f"Test  statistic - "
          f"Null: {tst_df[f'f_{i}'].isnull().mean():.2f} | "
          f"Mean: {tst_df[f'f_{i}'].mean():.2f} | Median {tst_df[f'f_{i}'].median():.2f} | "
          f"Min: {tst_df[f'f_{i}'].min():.2f} | Max {tst_df[f'f_{i}'].max():.2f} | "
          f"Unique num: {len(tst_df[f'f_{i}'].unique())}"
    )
    tst_ooc = not set(tst_df[f'f_{i}'].unique()).issubset(set(trn_df[f'f_{i}'].unique()))
    if tst_ooc:
        ooc_cat = set(tst_df[f'f_{i}'].unique()) - set(trn_df[f'f_{i}'].unique())
        sum_ = 0
        for k in ooc_cat:
              sum_ += (tst_df[f'f_{i}'] == k).sum()
        print(f'Test is Out-of-category. Num: {len(ooc_cat)} | Ratio: {sum_/len(tst_df[f"f_{i}"])}')        

    unused_cat = set(trn_df[f'f_{i}'].unique()) - set(tst_df[f'f_{i}'].unique())
    if len(unused_cat) > 0:
        sum_ = 0
        for k in unused_cat:
            sum_ += (trn_df[f'f_{i}'] == k).sum()
        print(f'Train exists unused category. Num: {len(unused_cat)} | Ratio: {sum_/len(trn_df[f"f_{i}"])}')
```

## Labels(is_clicked, is_installed)

```python
name = ['is_clicked', 'is_installed']
index = np.arange(len(name))

one = [trn_df['is_clicked'].mean(), trn_df['is_installed'].mean()]
zero = [1 - i for i in one]
plt.bar(index, zero, width=0.5, label='0')
plt.bar(index, one, width=0.5, bottom=zero, label= '1')
plt.ylim(0, 1)
plt.xticks(index, name)
plt.legend(loc='best',  shadow=True)

print(f"Train | is_clicked: {trn_df['is_clicked'].mean():.2f} | is_installed: {trn_df['is_installed'].mean():.2f}")
```

## Label Proportion of Each Day

```python
is_clicked = []
is_installed = []
for i in range(trn_df['f_1'].min(), trn_df['f_1'].max() + 1):
    tmp_df = trn_df[trn_df['f_1'] == i]
    is_clicked.append(tmp_df["is_clicked"].mean())
    is_installed.append(tmp_df["is_installed"].mean())
    print(f'Day {i} | is_clicked: {tmp_df["is_clicked"].mean():.2f} | is_installed: {tmp_df["is_installed"].mean():.2f}')
```

```python
day = [f'Day {i}' for i in range(45, 67)]
index = np.arange(len(day))
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

one = is_clicked
zero = [1 - i for i in one]
axs[0].bar(index, zero, width=0.4, label='0')
axs[0].bar(index, one, width=0.4, bottom=zero, label= '1')
axs[0].set_ylim(0, 1)
axs[0].set_xticklabels(day, rotation=90)
axs[0].legend(loc='best',  shadow=True)
axs[0].set_title('is_clicked')

one = is_installed
zero = [1 - i for i in one]
axs[1].bar(index, zero, width=0.4, label='0')
axs[1].bar(index, one, width=0.4, bottom=zero, label= '1')
axs[1].set_ylim(0, 1)
axs[1].set_xticklabels(day, rotation=90)
axs[1].legend(loc='best',  shadow=True)
axs[1].set_title('is_installed')


plt.show()
```
