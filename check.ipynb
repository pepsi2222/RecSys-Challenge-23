```python
ori_df = pd.read_csv('/root/autodl-tmp/xingmei/RecSys23/data/simplified_trn_val_tst_with_weekday.csv', sep='\t')
```

```python
trn_df = ori_df[0:3387880]
val_df = ori_df[3387880:-160973]
tst_df = ori_df[-160973:]
```

```python
trn_else = trn_df[(trn_df['f_1'] != 60) & (trn_df['f_1'] != 53)]
```

```python
trn_df_60 = ori_df[ori_df['f_1'] == 60]
trn_df_53 = ori_df[ori_df['f_1'] == 53]
split_row = int(trn_df_60.shape[0] * 0.5)
trn_df_60_1 = trn_df_60[0:split_row]
trn_df_60_2 = trn_df_60[split_row:]
split_row = int(trn_df_53.shape[0] * 0.5)
trn_df_53_1 = trn_df_53[0:split_row]
trn_df_53_2 = trn_df_53[split_row:]
split_row = int(val_df.shape[0] * 0.5)
val_df_1 = val_df[0:split_row]
val_df_2 = val_df[split_row:]
```

```python
trn_df = pd.concat([trn_else, trn_df_60_1, trn_df_53_1, val_df_1], axis=0)
val_df = pd.concat([trn_df_60_2, trn_df_53_2, val_df_2], axis=0)
all_df = pd.concat([trn_df, val_df, tst_df], axis=0)
```
